# Arbol de decision 

```python
# Train
model = tree.DecisionTreeClassifier()
model = model.fit(features[:,[21]], classes)

# Extract single tree
estimator = model

# Export as dot file
export_graphviz(estimator, out_file='decisionTree.dot', 
                feature_names = [features_names[21]] ,
                class_names = classes_names,
                rounded = True, proportion = False, 
                precision = 2, filled = True)

# Convert to png using system command (requires Graphviz)
os.system('dot -Tpng decisionTree.dot -o decisionTree.png -Gdpi=600')

# Display in jupyter notebook
Image(filename = 'decisionTree.png')
```
